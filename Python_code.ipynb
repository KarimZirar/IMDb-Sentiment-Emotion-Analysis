```python
# Dataset Preprocessing
import pandas as pd
import re
from langdetect import detect, LangDetectException
from tqdm import tqdm
import html

print("Loading data...")
reviews_df = pd.read_csv(r"C:\Users\kzira\Downloads\DATASET\selected_genres_merged.csv")

def clean_text(text):
    if not isinstance(text, str):
        return ""
    
    # Decode HTML first 
    text = html.unescape(text)
    
    # Remove ALL HTML tags 
    text = re.sub(r'<[^>]+>', ' ', text)
    
    # Remove usernames 
    text = re.sub(r'@\w+\s?', '', text)  # Now removes "@user123 " entirely
    
    # Standard cleaning
    text = re.sub(r'\\n', ' ', text)
    text = re.sub(r'URL: |http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

def is_english(text):
    if not isinstance(text, str) or not text.strip():
        return False
    try:
        return detect(text) == 'en'
    except LangDetectException:
        return False

print("Cleaning reviews...")
tqdm.pandas(desc="Reviews cleaned")
reviews_df["cleaned_review"] = reviews_df["review"].progress_apply(clean_text)

# Filtering out non-English reviews
print("Filtering English...")
reviews_df = reviews_df[reviews_df["cleaned_review"].apply(is_english)]

reviews_df.to_csv(r"C:\Users\kzira\Downloads\DATASET\selected_genres_cleaned.csv", index=False)
print(f"Cleaned {len(reviews_df)} reviews. Final checks:")

print("Remaining issues:")
print("- <br/> tags:", reviews_df["cleaned_review"].str.contains(r'<br/?>').sum())
print("- Usernames:", reviews_df["cleaned_review"].str.contains(r'@\w+').sum())

print("\nSample cleaned reviews:")
print(reviews_df[["review", "cleaned_review"]].sample(3).to_string())
```

```python
import pandas as pd
from transformers import pipeline
from collections import Counter
import torch
import gc

def load_models():
    try:
        device = 0 if torch.cuda.is_available() else -1
        
        # Load sentiment model 
        print("🔄 Loading sentiment model...")
        sentiment_model = pipeline(
            "text-classification",
            model="cardiffnlp/twitter-roberta-base-sentiment-latest",
            device=device,
            max_length=512,
            truncation=True
        )
        
        # Load emotion model with explicit config
        print("🔄 Loading emotion model...")
        emotion_model = pipeline(
            "text-classification",
            model="SamLowe/roberta-base-go_emotions",
            device=device,
            top_k=None, 
            max_length=512,
            truncation=True,
            function_to_apply="none"  
        )
        
        print("✅ Models loaded successfully!")
        return sentiment_model, emotion_model
        
    except Exception as e:
        print(f"❌ Model loading failed: {str(e)}")
        raise

def analyze_reviews(reviews, sentiment_model, emotion_model):
    sentiment_counts = Counter()
    emotion_counts = Counter()
    processed = 0
    
    for text in reviews:
        try:
            text = str(text).strip()[:512]
            if not text:
                continue
                
            # Sentiment analysis
            sentiment = sentiment_model(text)[0]
            sentiment_counts[sentiment['label']] += 1
            
            # Emotion analysis 
            emotions = emotion_model(text)[0]
            for emotion in emotions:
                if emotion['score'] > 0.1:  
                    emotion_counts[emotion['label']] += 1
            
            processed += 1
            if processed % 100 == 0:
                gc.collect()  
            
        except Exception as e:
            continue
            
    return sentiment_counts, emotion_counts, processed

try:
    sentiment_model, emotion_model = load_models()
    
    df = pd.read_csv(r"C:\Users\kzira\Downloads\DATASET\selected_genres_cleaned_FINAL.csv")
    print(f"✅ Loaded {len(df)} reviews")
    
    # Analyze movies
    movies = [
        ("Deadpool", 2016),
        ("The Conjuring", 2013),
        ("The Green Mile", 1999),
        ("The Mist", 2007)
    ]
    
    results = {}
    for movie, year in movies:
        print(f"\n🎬 Analyzing {movie} ({year})...")
        
        reviews = df[
            (df['movie_title'].str.strip().str.lower() == movie.strip().lower()) & 
            (df['year'] == year)
        ]['cleaned_review'].dropna()
        
        if len(reviews) == 0:
            print("⚠️ No reviews found")
            continue
            
        sentiment_counts, emotion_counts, processed = analyze_reviews(
            reviews, sentiment_model, emotion_model
        )
        
        if processed == 0:
            print("⚠️ No valid reviews processed")
            continue
            
        # Calculate results
        results[movie] = {
            'total': processed,
            'sentiment': {
                'counts': dict(sentiment_counts),
                'percentages': {k: f"{v/processed:.1%}" for k,v in sentiment_counts.items()}
            },
            'emotions': {
                'counts': dict(emotion_counts.most_common(5)),
                'percentages': {k: f"{v/sum(emotion_counts.values()):.1%}" 
                              for k,v in emotion_counts.most_common(5)}
            }
        }
        print(f"✅ Processed {processed} reviews")
    
    print("\n RESULTS")
    for movie, data in results.items():
        print(f"\n📽 {movie.upper()} ({data['total']} reviews)")
        print("\nSentiment:")
        for sent in ['positive', 'neutral', 'negative']:
            print(f"- {sent}: {data['sentiment']['counts'].get(sent,0)} ({data['sentiment']['percentages'].get(sent,'0%')})")
        
        print("\nTop Emotions:")
        for emo, count in data['emotions']['counts'].items():
            print(f"- {emo}: {count} ({data['emotions']['percentages'][emo]})")
        print("━"*50)
        
except Exception as e:
    print(f"error: {str(e)}")
```

```python
import pandas as pd

# Get the exact movie titles from your data
all_titles = combined_results["movie_title"].str.strip().str.lower().unique()

# Map our target movies to their actual titles in the data
title_mapping = {
    "deadpool": None,
    "the conjuring": None, 
    "the green mile": None,
    "the mist": None
}

# Find matching titles
for title in all_titles:
    lower_title = title.lower()
    for target in title_mapping:
        if target in lower_title:
            title_mapping[target] = title

print("Discovered title mappings:")
for orig, actual in title_mapping.items():
    print(f"{orig:15} => {actual}")

# Calculate confidence scores using properly matched titles
confidence_by_movie = []

for target_name, actual_name in title_mapping.items():
    if actual_name is None:
        print(f"\nWarning: No match found for '{target_name}'")
        continue
        
    try:
        movie_reviews = combined_results[
            combined_results["movie_title"].str.strip() == actual_name
        ]
        
        if len(movie_reviews) == 0:
            print(f"No reviews found for '{actual_name}'")
            continue
            
        avg_sentiment = movie_reviews["sentiment_score"].mean()
        avg_emotion = movie_reviews["top_emotion_score"].mean()
        
        confidence_by_movie.append({
            "Movie": actual_name,
            "Reviews": len(movie_reviews),
            "Sentiment Conf": round(avg_sentiment, 2),
            "Emotion Conf": round(avg_emotion, 2)
        })
        
    except Exception as e:
        print(f"Error processing {actual_name}: {str(e)}")
        continue

# Create final table
confidence_df = pd.DataFrame(confidence_by_movie)
print("\nFinal Confidence Scores:")
print(confidence_df.to_string(index=False))
```

```python
# Sentiment/emotion analysis across genres
import pandas as pd
import torch
from transformers import pipeline
from collections import Counter
import gc
from tqdm import tqdm


df = pd.read_csv(r"C:\Users\kzira\Downloads\DATASET\selected_genres_cleaned_FINAL.csv")
print(f"Loaded {len(df)} reviews from the CSV file.")

device = 0 if torch.cuda.is_available() else -1

sentiment_model = pipeline(
    "text-classification",
    model="cardiffnlp/twitter-roberta-base-sentiment-latest",
    device=device,
    max_length=512,
    truncation=True
)
emotion_model = pipeline(
    "text-classification",
    model="SamLowe/roberta-base-go_emotions",
    device=device,
    top_k=None,
    max_length=512,
    truncation=True,
    function_to_apply="none"
)

# Batch Analysis Function for a Genre
def analyze_genre_reviews(reviews, batch_size=32):
    """
    Processes reviews in batches to compute overall sentiment and emotion distributions.
    Returns:
      sentiment_counts: Counter with counts of each sentiment label.
      emotion_counts: Counter with counts of each emotion label.
    """
    sentiment_counts = Counter()
    emotion_counts = Counter()
    num_reviews = len(reviews)

    for idx in tqdm(range(0, num_reviews, batch_size), desc="Processing batches"):
        batch = reviews[idx: idx + batch_size].tolist()
        
        try:
            sentiments = sentiment_model(batch)
            for result in sentiments:
                sentiment_counts[result['label']] += 1
        except Exception as e:
            print(f"Sentiment batch error: {e}")
        
        try:
            emotions_results = emotion_model(batch)
            for result in emotions_results:
                for emotion in result:
                    if emotion['score'] > 0.1:  
                        emotion_counts[emotion['label']] += 1
        except Exception as e:
            print(f"Emotion batch error: {e}")
        
        gc.collect()
        if device >= 0:
            torch.cuda.empty_cache()
    
    return sentiment_counts, emotion_counts

# Sample Test on a Small Subset for 'Drama'
drama_reviews_sample = df[df['genre'].str.lower() == 'drama']['cleaned_review'].dropna().head(100)
sample_sentiments, sample_emotions = analyze_genre_reviews(drama_reviews_sample, batch_size=32)
print("Sample Sentiment Results:", sample_sentiments)
print("Sample Emotion Results:", sample_emotions)

genres = {
    "Drama": 181208,
    "Thriller": 110823,
    "Horror": 102551,
    "Comedy": 83927,
    "Romance": 75861
}

results = {}
for genre in genres.keys():
    print(f"\nProcessing genre: {genre}")
    genre_reviews = df[df['genre'].str.lower() == genre.lower()]['cleaned_review'].dropna()
    
    if len(genre_reviews) > 5120:
        genre_reviews = genre_reviews.sample(n=5120, random_state=42)
    
    s_counts, e_counts = analyze_genre_reviews(genre_reviews, batch_size=32)
    
    total_sentiments = sum(s_counts.values())
    sentiment_percentages = {label: f"{count / total_sentiments:.1%}" for label, count in s_counts.items()}
    
    total_emotions = sum(e_counts.values())
    top_emotions = e_counts.most_common(5)
    top_emotion_counts = {label: count for label, count in top_emotions}
    emotion_percentages = {label: f"{count / total_emotions:.1%}" for label, count in top_emotions}
    
    results[genre] = {
        'sentiments': {
            'counts': dict(s_counts),
            'percentages': sentiment_percentages
        },
        'emotions': {
            'counts': top_emotion_counts,
            'percentages': emotion_percentages
        }
    }

# Preview the Results
for genre, data in results.items():
    print(f"\nGenre: {genre}")
    print("Sentiment Distribution:")
    for label, count in data['sentiments']['counts'].items():
        print(f"  - {label}: {count} ({data['sentiments']['percentages'][label]})")
    print("Top 5 Emotion Distribution:")
    for label, count in data['emotions']['counts'].items():
        print(f"  - {label}: {count} ({data['emotions']['percentages'][label]})")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

emotion_data = {
    'Drama': {'Admiration': 68.2, 'Disappointment': 6.9, 'Neutral': 5.8, 'Love': 3.9, 'Joy': 3.1},
    'Thriller': {'Admiration': 65.1, 'Disappointment': 8.9, 'Neutral': 6.7, 'Joy': 3.3, 'Love': 2.3},
    'Horror': {'Admiration': 53.9, 'Disappointment': 10.0, 'Neutral': 5.1, 'Joy': 5.0, 'Fear': 4.2},
    'Comedy': {'Admiration': 58.8, 'Amusement': 9.8, 'Disappointment': 6.7, 'Joy': 6.2, 'Love': 5.2},
    'Romance': {'Admiration': 63.8, 'Disappointment': 7.0, 'Love': 6.7, 'Joy': 4.6, 'Neutral': 4.2},
}

emotion_colors = {
    'Admiration': "#3498db",  
    'Disappointment': "#f39c12",  
    'Neutral': "#95a5a6",  
    'Love': "#e74c3c",  
    'Joy': "#2ecc71",  
    'Fear': "#8e44ad",  
    'Amusement': "#9b59b6"  
}

def plot_emotion_distribution(genre, emotions, percentages):
    plt.figure(figsize=(10, 5))
    palette = [emotion_colors[emotion] for emotion in emotions]  
    sns.barplot(x=percentages, y=emotions, palette=palette, edgecolor="black")
    
    plt.title(f"Top Emotions in {genre} Movies", fontsize=14, pad=20)
    plt.xlabel("Percentage of Reviews (%)", fontsize=12)
    plt.xlim(0, 80)  
    plt.ylabel("")
    sns.despine(left=True)  
    for i, pct in enumerate(percentages):
        plt.text(pct + 1, i, f"{pct}%", va="center", fontsize=11)
    
    plt.tight_layout()
    plt.show()

for genre, data in emotion_data.items():
    emotions = list(data.keys())
    percentages = list(data.values())
    plot_emotion_distribution(genre, emotions, percentages)

```
